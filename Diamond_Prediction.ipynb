```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dia=pd.read_csv("diamonds.csv")
```

```python
dia.head()
```

```python
dia.columns
```

```python
def impute_col(cols):
    if cols=='D':
        return 6
    if cols=='E':
        return 5
    if cols=='F':
        return 4
    if cols=='G':
        return 3
    if cols=='H':
        return 2
    if cols=='I':
        return 1
    else:
        return 0
def impute_clr(cols):
    if cols=='IF':
        return 7
    if cols=='VVS1':
        return 6
    if cols=='VVS2':
        return 5
    if cols=='VS1':
        return 4
    if cols=='VS2':
        return 3
    if cols=='SI1':
        return 2
    if cols=='SI2':
        return 1
    else:
        return 0
def impute_cut(cols):
    if cols=='Ideal':
        return 4
    if cols=='Premium':
        return 3
    if cols=='Very Good':
        return 2
    if cols=='Good':
        return 1
    else:
        return 0 
```

```python
dia['color']=dia['color'].apply(impute_col)
dia['clarity']=dia['clarity'].apply(impute_clr)
dia['cut']=dia['cut'].apply(impute_cut)
```

```python
dia
```

```python
sns.set_style("whitegrid")
```

```python
sns.pairplot(dia)
```

```python
dia.columns
```

```python
x=dia[['cut', 'color', 'clarity','x', 'y', 'z','price','depth','table']]
y=dia['carat']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.5,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print("Coefficients:")
print(lm.coef_)
```

```python
predictions=lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
print("MAE: ", metrics.mean_absolute_error(y_test,predictions))
print("MSE: ", metrics.mean_squared_error(y_test,predictions))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
sns.distplot(y_test-predictions,bins=50)
```
